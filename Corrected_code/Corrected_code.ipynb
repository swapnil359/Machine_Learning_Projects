```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import train_test_split

# Assuming X_train, X_test, y_train, y_test are already defined as part of the data preprocessing step

# Define models (excluding XGBoost)
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}

# To store the results of each model
results = {}

# Loop through models
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # Get probabilities for ROC AUC
    
    # Evaluate performance
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    
    # Store results
    results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "AUC Score": auc
    }

# Print the results
for model_name, metrics in results.items():
    print(f"Model: {model_name}")
    for metric, score in metrics.items():
        print(f"{metric}: {score:.4f}")
    print("\n")
```

```python
import pandas as pd

# Create a DataFrame with actual and predicted values
results_df = pd.DataFrame({'Actual': y_test, 'Predicted_Probability': y_proba})
```

```python
# Sort by predicted probabilities
results_df = results_df.sort_values('Predicted_Probability')

# Calculate cumulative sums for each class
total_default = results_df['Actual'].sum()
total_non_default = len(results_df) - total_default

# Calculate cumulative distribution for default
results_df['Cumulative_Default'] = results_df['Actual'].cumsum() / total_default
# Calculate cumulative distribution for non-default
results_df['Cumulative_Non_Default'] = (1 - results_df['Actual']).cumsum() / total_non_default
```

```python
# Calculate K-S statistic
results_df['KS'] = results_df['Cumulative_Default'] - results_df['Cumulative_Non_Default']
ks_statistic = results_df['KS'].max()
print(f"K-S Statistic: {ks_statistic}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(results_df['Predicted_Probability'], results_df['Cumulative_Default'], label='Cumulative Default', color='blue')
plt.plot(results_df['Predicted_Probability'], results_df['Cumulative_Non_Default'], label='Cumulative Non-Default', color='orange')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(x=results_df.loc[results_df['KS'].idxmax(), 'Predicted_Probability'], color='red', ls='--', label='K-S Point')
plt.title('K-S Chart for Default Prediction')
plt.xlabel('Predicted Probability of Default')
plt.ylabel('Cumulative Distribution')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with predicted probabilities and actual labels
ks_df = pd.DataFrame({'Actual': y_test, 'Probability': y_proba})

# Sort by predicted probabilities
ks_df = ks_df.sort_values(by='Probability')

# Calculate the cumulative distribution for both classes
ks_df['Cum_Pos'] = np.cumsum(ks_df['Actual']) / ks_df['Actual'].sum()
ks_df['Cum_Neg'] = np.cumsum(1 - ks_df['Actual']) / (len(y_test) - ks_df['Actual'].sum())

# Calculate the K-S statistic
ks_statistic = np.max(np.abs(ks_df['Cum_Pos'] - ks_df['Cum_Neg']))
print(f"K-S Statistic: {ks_statistic}")

# Plot K-S Chart
plt.figure(figsize=(10, 6))
plt.plot(ks_df['Probability'], ks_df['Cum_Pos'], label='Cumulative Positive', color='blue')
plt.plot(ks_df['Probability'], ks_df['Cum_Neg'], label='Cumulative Negative', color='red')
plt.title('K-S Chart')
plt.xlabel('Predicted Probability')
plt.ylabel('Cumulative Proportion')
plt.legend()
plt.axhline(ks_statistic, color='green', linestyle='--', label=f'K-S Statistic: {ks_statistic:.4f}')
plt.show()
```
